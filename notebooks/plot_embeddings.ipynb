```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
model_type = 'n'
```

```python
d_fire_embeddings = np.load(f'../data/embeddings/yolo_{model_type}_d_fire_embeddings.npy')
finland_embeddings = np.load(f'../data/embeddings/yolo_{model_type}_finland_embeddings.npy')
flame2_embeddings = np.load(f'../data/embeddings/yolo_{model_type}_flame2_embeddings.npy')

d_fire_labels = np.load(f'../data/labels/yolo_d_fire_labels.npy')[:, 1]
finland_labels = np.load(f'../data/labels/yolo_finland_labels.npy')[:, 0]
flame2_labels = np.load(f'../data/labels/yolo_flame2_labels.npy')[:, 1]

# Reduce dimensionality to 2D using PCA
pca = PCA(n_components=2)
d_fire_embeddings_2d = pca.fit_transform(d_fire_embeddings)
finland_embeddings_2d = pca.transform(finland_embeddings)
flame2_embeddings_2d = pca.transform(flame2_embeddings)

# Plot the distribution
plt.figure(figsize=(8, 6))
plt.scatter(d_fire_embeddings_2d[:, 0], d_fire_embeddings_2d[:, 1], alpha=0.7)
plt.scatter(finland_embeddings_2d[:, 0], finland_embeddings_2d[:, 1], alpha=0.7)
plt.scatter(flame2_embeddings_2d[:, 0], flame2_embeddings_2d[:, 1], alpha=0.7)
plt.title(f'2D t-SNE of YOLO Embeddings ({model_type})')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(['D-Fire', 'Finland', 'Flame2'])
plt.grid(True)
plt.show()
```

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
d_fire_embeddings_2d = tsne.fit_transform(d_fire_embeddings)
finland_embeddings_2d = tsne.fit_transform(finland_embeddings)
flame2_embeddings_2d = tsne.fit_transform(flame2_embeddings)

# Plot the distribution
plt.figure(figsize=(8, 6))
plt.scatter(d_fire_embeddings_2d[:, 0], d_fire_embeddings_2d[:, 1], alpha=0.7)
plt.scatter(finland_embeddings_2d[:, 0], finland_embeddings_2d[:, 1], alpha=0.7)
plt.scatter(flame2_embeddings_2d[:, 0], flame2_embeddings_2d[:, 1], alpha=0.7)
plt.title(f'2D t-SNE of YOLO Embeddings ({model_type})')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend(['D-Fire', 'Finland', 'Flame2'])
plt.grid(True)
plt.show()

plt.savefig(f'../images/tsne_yolo_{model_type}.png', bbox_inches='tight', dpi=200)
plt.show()
```
